```python
# CSV and StringIO
from io import StringIO, BytesIO
```

```python
# Declaring data which is equivalent to csv data
data = ('x,y,z,a\n'
       '1,2,3,4\n'
       '5,6,7,8\n'
       '8,9,10,11')
data
```

```python
type(data)
```

```python
# converting sting data to padas dataframe
data_io = pd.read_csv(StringIO(data))
print(data_io)
type(data_io)
```

```python
# Extracting data from specific columns and converting into csv file

data_io = pd.read_csv(StringIO(data), usecols= ['x','z'])
print(data_io)
data_io.to_csv('data.csv')
print("Exported extracted data successfully into csv")
```

```python
# Changing datatypes of the columns

data = ('x,y,z,a\n'
       '1,2,3,4\n'
       '5,6,7,8\n'
       '8,9,10,11')
print(data)
data = pd.read_csv(StringIO(data), dtype = {'x':int, 'y':float, 'z':'Int64', 'a':object })
print (data)

# Filtering the columns
print('{} and its type = {}'.format(data['x'][0],type(data['x'][0])))
print('{} and its type = {}'.format(data['y'][1],type(data['y'][1])))
print('{} and its type = {}'.format(data['z'][2],type(data['z'][2])))
print('{} and its type = {}'.format(data['a'][2],type(data['a'][2])))
```

```python
data = ('x,y,z,a\n'
       '1,2,"car",4\n'
       '5,6,"apple",8\n'
       '8,9,"bat"')

data = pd.read_csv(StringIO(data))
data
```

```python
# index_cols

data = ('x,y,z,a\n'
       '1,2,"car",4\n'
       '5,6,"apple",8\n'
       '8,9,"bat"')

# using index_cols
data = pd.read_csv(StringIO(data), index_col='z') # here now column 'z' became the index column; we can use the name/index number
print(data)
```

```python
data = ('a,b,c\n'
        'cat1,apple,bat,\n'
        'cat2,orange,cow,') # header has 3 cols but data has 4 cols

print(pd.read_csv(StringIO(data)))
```

```python
data = ('a,b,c\n'
        'cat1,apple,bat,\n'
        'cat2,orange,cow,') # header has 3 cols but data has 4 cols

print(pd.read_csv(StringIO(data), index_col=False)) # Use index_col=False to set the columns formatting
```

```python
# Combining usecols and index_cols

data = ('a,b,c\n'
        'cat1,apple,bat,\n'
        'cat2,orange,cow,') # header has 3 cols but data has 4 cols

df1 = pd.read_csv(StringIO(data), index_col=False, usecols = ['a','c'])
df1
```

```python
## Quoting and Escape CharactersÂ¶. Very useful in NLP

data = 'a,b\n"Hello, \\"Bob\\", nice to see you",5'
pd.read_csv(StringIO(data), escapechar='\\')
```
